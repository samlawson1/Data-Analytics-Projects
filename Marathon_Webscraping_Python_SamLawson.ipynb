<font color = blue>**Instead of writing a for loop for every time I need to create a dataframe, I now want to create a function that will do that for me when I call it. To do that, I make a few modifications to my existing for loop and nest it inside a function.**</font>

```python
def create_df(url_base, page_count, df):
    for page in range(1, page_count + 1):
        url = url_base + str(page)
        #print(url)
        r = requests.post(url)
        #print(r.status_code)
        soup = BeautifulSoup(r.text, 'lxml')
        tables = soup.find_all("table", attrs = {"class":"table table-responsive table-bordered"})
        results = pd.read_html(str(tables[0])) 
        df = df.append(results)
    return df.reset_index(drop = True)
```
